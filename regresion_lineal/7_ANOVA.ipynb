```python
# Tratamiento de datos
# -----------------------------------------------------------------------
import numpy as np
import pandas as pd
import random 

# Gr√°ficos
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Estad√≠sticos
# ------------------------------------------------------------------------------
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
from sklearn.preprocessing import StandardScaler

plt.rcParams["figure.figsize"] = (10,8) 
```

```python
df = pd.read_csv('../datos/supermarket_no_estandarizado.csv', index_col = 0)
df.head(1)
```

```python
columnas_nuevas = {col: col.replace(' ', '_').replace('%', '') for col in df.columns}
```

```python
df.rename(columns = columnas_nuevas, inplace = True)
```

```python
df.head(1)
```

```python
lm = ols('Rating ~ Quantity  + City + Tax_5 +  Total + Unit_price + Customer_type + Gender + Payment + hora_compra + mes_compra' , data=df).fit()
sm.stats.anova_lm(lm)
```

```python
lm2 = ols('RATING_BOX ~ Quantity  + City + Tax_5 +  Total + Unit_price + Customer_type + Gender + Payment + hora_compra + mes_compra' , data=df).fit()
sm.stats.anova_lm(lm2)
```

```python
lm2.summary()
```

```python
lm.summary()
```

Tras chequear los resultados del ANOVA, comprobamos que la columna de ciudad que no hab√≠amos considerado hasta ahora, SI puede tener relaci√≥n en el Rating del cliente. Las variables num√©ricas como ya hab√≠amos visto no tienen ning√∫n efecto, y las categ√≥ricas al contrario de lo que pens√°bamos tampoco tienen ningun efecto.

Vemos que tenemos un R-squared de 0.016: el resultado es muy bajo, por lo que concluimos que las variables predictoras no explican nuestra variable respuesta ('Rating').
